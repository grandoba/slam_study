```python
from numpy.random import randn
```

```python
class PosSensor(object):
    def __init__(self, pos=[0,0], vel=[0,0], noise_std=1, freq=1):
        self.vel = vel
        self.noise_std = noise_std/freq
        self.pos = pos # [pos[0], pos[1]]
        self.dt = 1/freq
    def read(self):
        self.pos[0] += self.vel[0]*self.dt
        self.pos[1] += self.vel[1]*self.dt
        return [self.pos[0] + randn() * self.noise_std,
                self.pos[1] + randn() * self.noise_std]
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
pos, vel = [4,3],[2,1]
sensor = PosSensor(pos,vel,noise_std=1,freq=1)
ps = np.array([sensor.read() for _ in range(50)])
plt.autoscale(tight=False)
plt.scatter(ps[:,0],ps[:,1])
```
